```python
import pandas as pd
import json
```

```python
# Функция для загрузки данных из JSON файла в DataFrame
def load_json_to_dataframe(file_name, column_name):
    with open(file_name, 'r') as file:
        data = json.load(file)
        df = pd.DataFrame(data[column_name])
        df.columns = ['date', column_name]
        df['date'] = pd.to_datetime(df['date'], unit='ms')
        return df
```

```python
# Словарь файлов и соответствующих им DataFrame
file_mapping = {
    'miners-revenue.json': 'miners-revenue'
}
```

```python
# Создание и заполнение словаря DataFrame
dataframes = {}
for file, column in file_mapping.items():
    dataframes[column] = load_json_to_dataframe(file, column)
    print(dataframes[column])
```

```python
revenue = dataframes['miners-revenue']
revenue.to_csv('revenue.csv', index=False)
```

```python
from functools import reduce

# Список всех DataFrame
dfs = list(dataframes.values())

# Объединение всех DataFrame по столбцу 'date'
merged_df = reduce(lambda left, right: pd.merge(left, right, on='date', how='outer'), dfs)

# Вывод объединенного DataFrame
merged_df
```

```python
# Замена NaN значений на средние
merged_df.fillna(merged_df.mean(), inplace=True)

# Вывод обновленного DataFrame
merged_df
```

`Теперь оценим корреляцию между исследуемыми признаками`

```python
# Вычисление матрицы корреляции
correlation_matrix = merged_df.corr()

import seaborn as sns
import matplotlib.pyplot as plt

# Создание тепловой карты
plt.figure(figsize=(10, 8))
heatmap = sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Матрица корреляции')
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd

btc_dataframe = pd.read_csv('btc.csv')
btc = btc_dataframe
date_column = btc_dataframe['date']
df_without_date = btc_dataframe.drop('date', axis=1)


# Создание объекта MinMaxScaler
scaler = MinMaxScaler()

# Нормализация всех столбцов в dataframe df
btc_df_norm = pd.DataFrame(scaler.fit_transform(df_without_date), columns=df_without_date.columns)

btc_dataframe = pd.concat([date_column, btc_df_norm], axis=1)


btc_dataframe.fillna(btc_dataframe.mean(), inplace=True)


plt.plot(btc_dataframe['miners-revenue'], marker='.', label='BTC')
plt.title('Доходность (BTC)')
plt.ylabel('Revenue')
plt.xlabel('Time')
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt

# Преобразование датасета
scaler = MinMaxScaler(feature_range=(0, 1))

scaled_data = scaler.fit_transform(btc[['miners-revenue']])

# Создание временных окон
def create_dataset(X, time_steps=1):
    Xs, ys = [], []
    for i in range(len(X) - time_steps):
        Xs.append(X[i:(i + time_steps)])
        ys.append(X[i + time_steps])
    return np.array(Xs), np.array(ys)

TIME_STEPS = 30
X, y = create_dataset(scaled_data, TIME_STEPS)

# Разделение датасета на обучающую и тестовую выборки
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Создание и обучение модели
model = Sequential()
model.add(LSTM(64, input_shape=(X_train.shape[1], X_train.shape[2])))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
history = model.fit(X_train, y_train, epochs=40, batch_size=32, validation_data=(X_test, y_test), verbose=1, shuffle=False)
```

```python
# Предсказание на тестовой выборке
y_pred = model.predict(X_test)

# Построение графика
plt.plot(y_test, marker='.', label='Истинная')
plt.plot(y_pred, 'r', marker='.', label='Предсказанная')
plt.title('Доходность (BTC)')
plt.ylabel('Доходность')
plt.xlabel('Время')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Оценка модели с использованием различных метрик
y_pred = model.predict(X_test)

# Mean Squared Error (MSE)
mse = mean_squared_error(y_test, y_pred)
print('Mean Squared Error:', mse)

# Mean Absolute Error (MAE)
mae = mean_absolute_error(y_test, y_pred)
print('Mean Absolute Error:', mae)

# Coefficient of Determination (R-squared)
r2 = r2_score(y_test, y_pred)
print('R-squared:', r2)
```
